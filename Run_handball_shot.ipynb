# Evaluation

```python
import os
import json
import numpy as np


def mpjpe(pred, gt):
    error = np.linalg.norm(pred - gt, ord=2, axis=-1).mean()
    return error


def VIM(pred, GT, calc_per_frame=True, return_last=True):
    if calc_per_frame:
        pred = pred.reshape(-1, 39)
        GT = GT.reshape(-1, 39)
    errorPose = np.power(GT - pred, 2)
    errorPose = np.sum(errorPose, 1)
    errorPose = np.sqrt(errorPose)
    
    if return_last:
        errorPose = errorPose[-1]
    return errorPose

y_test = np.load("./data/handball_shot/test.npy")[:, :, -14:]
```

```python
# Trained on 3DPW
d = {}
for name in ["handball_shot_"+f+".npy" for f in ["zero_velocity", "ltd", "dvita", "tbiformer", "mrt", "socialtgcn", "jrtransformer", "mpfsir", "future_motion_iccv21", "somoformer"]]:
    d[name.split(".npy")[0]] = np.load("./data/predictions/" + name).reshape(-1, 2, 14, 13, 3)
d_pt = d
```

```python
print("***** VIM *****")
print("Model_name                          | 75ms  | 175ms | 375ms  | 475ms  | 675ms  | AVG")
for modelname, y_pred in d.items():
    vims = [ np.mean( [(VIM(pred[0][:LEN], gt[0][:LEN]) + VIM(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(y_pred, y_test)] ) * 100 for LEN in [2, 4, 8, 10, 14]]
    print("{0} | {1:.2f} | {2:.2f} | {3:.2f} | {4:.2f} | {5:.2f} | {6:.2f}".format(modelname.ljust(35), *vims, np.mean(vims)))
  
```

```python
print("***** MPJPE *****")
print("Model_name                          | 75ms   | 175ms  | 375ms  | 475ms  | 675ms  | AVG")
for modelname, y_pred in d.items():
    errs = [ np.mean( [(mpjpe(pred[0][:LEN], gt[0][:LEN]) + mpjpe(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(y_pred, y_test)] ) * 1000 for LEN in [2, 4, 8, 10, 14]]    
    print("{0} | {1:.2f} | {2:.2f} | {3:.2f} | {4:.2f} | {5:.2f} | {6:.2f}".format(modelname.ljust(35), *errs, np.mean(errs)))
  
```

```python
print("***** MPJPE *****")
print("Model_name                          | 75ms   | 175ms  | 375ms  | 475ms  | 675ms  | AVG")
for modelname, y_pred in d.items():
    errs = [ np.mean( [(mpjpe(pred[0][:LEN], gt[0][:LEN]) + mpjpe(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(y_pred, y_test)] ) * 1000 for LEN in [2, 4, 8, 10, 14]]    
    print("{0} | {1:.2f} | {2:.2f} | {3:.2f} | {4:.2f} | {5:.2f} | {6:.2f}".format(modelname.ljust(35), *errs, np.mean(errs)))
  
```

```python
# Relative perfomance increase (or error decrease)
```

```python
print("***VIM***")
for k in d.keys():

    pt_vim = np.mean([ np.mean( [(VIM(pred[0][:LEN], gt[0][:LEN]) + VIM(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(d_pt[k], y_test)] ) * 100 for LEN in [2, 4, 8, 10, 14]])
    
    ft_vim = np.mean([ np.mean( [(VIM(pred[0][:LEN], gt[0][:LEN]) + VIM(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(d[k], y_test)] ) * 100 for LEN in [2, 4, 8, 10, 14]])
    
    print("{0} {1:.2f}%".format(k.ljust(35), ((pt_vim - ft_vim) / pt_vim) * 100))


print("\n\n***MPJPE***")
for k in d.keys():
    
    pt_mpjpe = np.mean([ np.mean( [(mpjpe(pred[0][:LEN], gt[0][:LEN]) + mpjpe(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(d_pt[k], y_test)] ) * 1000 for LEN in [2, 4, 8, 10, 14]])
    
    ft_mpjpe = np.mean([ np.mean( [(mpjpe(pred[0][:LEN], gt[0][:LEN]) + mpjpe(pred[1][:LEN], gt[1][:LEN])) / 2. for pred, gt in zip(d[k], y_test)] ) * 1000 for LEN in [2, 4, 8, 10, 14]])
    
    print("{0} {1:.2f}%".format(k.ljust(35), ((pt_mpjpe - ft_mpjpe) / pt_mpjpe) * 100))
    
```
